```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://worldpopulationreview.com/country-rankings/developed-countries"
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text,'html')
print(soup)
```

```python
table = soup.find('table')
print(table)
```

```python
Headings = table.find_all('th')
print(Headings)
```

```python
Header = [i.text for i in Headings]
print(Header)
```

```python
df = pd.DataFrame(columns = Header)
df
```

```python
rows = table.find_all('tr')
print(rows)
```

```python
ind=0
for i in rows[1:11]:
  row_data = i.find_all('td')
  every_row_data = [i.text for i in row_data]
  print(every_row_data)
  df.loc[ind] = every_row_data
  ind = ind+1
```

```python
df
```

```python
df.to_csv("Top 10 Developed Countries (2021 HDI).csv")
```
